## Fase II: Campanhas de Investigação com análise de desempenho do Wi-Fi

### Hands-On 04: Campanha 03
### Cenário 02: Posição Aleátoria

O objetivo da campanha 03 é a avaliação dos algoritmos de adaptação de taxa com o aumento do número de estações (STAs), em que foram estudados três tipos de cenários, um mantendo a posição das STAs constante, outro variando a posição aleatoriamente e um terceiro cenário variando a posição com o acréscimo de mobilidade , segue 
abaixo os parâmetros especificados para realizar as simulações :
* __Padrão escolhido__ : 802.11ac
* __Largura de canal__: 80MHz;
* __Frequência de operação__: 5GHz;
* __Algoritmos Utilizados__ : MinstrelHtWifiManager, ConstantRateWifiManager, IdealWifiManager;
* __Taxa oferecida__: 1.5Mbps (downlink) e 750kbps (uplink);
* __Número de STAs__: 1, 2, 3, 4, 5, 6, 7, 8, 9 e 10;
* __Tempo de simulação__: 500 segundos;
* __Simulation Runs__: 1000 

Com isso foram feitas modificações no codigo [rate-adaptation-distance.cc](https://www.nsnam.org/docs/release/3.29/doxygen/rate-adaptation-distance_8cc.html), disponibilizado nos exemplos wireless do ns-3 (~/ns-3.29/examples/wireless), para o cenário especiado variação da posição das STAs aleatoriamente, e partir daí como explicado no  [Hands-On 1.5: FlowMonitor, campanhas e plotagem de gráficos](https://nbviewer.jupyter.org/github/vicentesousa/ns3-intermediate-training/blob/master/fase_02/HD_01-5/f02_hd01-5.ipynb), a inserção do flow monitor no código para conseguir visualizar as métricas da rede, foram criadas as campanha para 1000 JOBS e simuladas no Núcleo de Processamento de Alto Desempenho - NPAD, os arquivos obtidos resultado da simulação foram utilizados para plotagem dos gráficos no Matlab, sendo os gráficos do __Throughput__, __ Packet Loss Ratio__ e  __User Satisfaction__ em função do número de STAs.


## Resultados 
#### Downlink Throughput
Começamos a análise dos resultados com o gráfico do Throughput em função do número de STAs que pode ser visto abaixo:

![fig_figura](./FIGS/Aleatoria/T-put_dl.png)

Pode ser observado que a taxa inicial oferecida está dentro do esperado de 1.5 Mbps, para os algoritmos  MinstrelHtWifiManager e para o ConstanteRateWifiManager, a partir daí devido as característica de se adpatar ao canal do algoritmo MinstrelHtWifiManager manteve sua taxa recebida quase constante com um suave declinio com o aumento do número de estações, já o ConstanteRateWifiManager a partir de duas estações teve um decaimento alto chegando ao valor muito baixo de taxa, o comportamento do ideal seguiu inconclusivo durante a simulação.

#### Uplink Throughput
A análise para o uplink a taxa incial oferecida também está dentro do esperado de 750 kbps,o algortimo MINSTRELHTwifiManeger manteve seu comportamento com uma taxa com pouca variação da inicial, matendo um bom desempenho,o ConstanteRateWifiManager obteve uma melhora em seu desempenho decaindo a partir da 4 estação,chegando um taxa mínima de aproximadamente 350 kbps.

![fig_figura](./FIGS/Aleatoria/T-put_ul.png) 

#### Downlink Packet Loss Ratio

A próxima análise é em relação entre a perda de pacotes, em que o desempenho dos algoritmos foi semelhante as análises anteriores, tendo o MinstrelHtWifiManager com o melhor desempenho devido a baixa perda de de pacotes  independente do numero de STAs, já o ConstanteRateWifiManager apresenta um bom resultado até as duas primeiras STAs a partir daí sofre um aumento considerável, até chegar aos aproximadamente 90% de perda com as 10 STAs.

![fig_figura](./FIGS/Aleatoria/PlR_dl.png)


#### Uplink Packet Loss Ratio

Para uplink a perda de pacotes manteve o mesmo comportamento para o MinstrelHtWifiManager e para o algoritmo ConstanteRateWifiManager apesar de apresentar um baixo desempenho obteve uma melhora considerável tendo uma perda máxima de 16%.

![fig_figura](./FIGS/Aleatoria/PLR_ul.png)

####  Downlink  User Satisfaction

A análise é para User staisfaction com o critério de 10% da taxa observamos novamente o MinstrelHtWifiManager apresenta um bom desempenho, quase 100% de satisfação do usuário para quase todas as STAs, já o ConstanteRateWifiManager como no decorrer de todas as análises apresenta um bom desempenho inicial mais assim que é aumentado o numero de estações começa a decair tendo um nível mínimo de satisfação do usuário de quase 65%.

![fig_figura](./FIGS/Aleatoria/Us_dl.png)

#### Uplink  User Satisfaction

Para uplink a mesma análise do dowlink pode ser feita, o comportamento MinstrelHtWifiManager não muda, e o do ConstanteRateWifiManager sofre uma pequena melhora começando seu decaimento em 4 STAs em vez de 3.

![fig_figura](./FIGS/Aleatoria/Us_ul.png)
