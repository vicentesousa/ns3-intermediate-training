### Função de Coordenação Distribuída(DCF)

### Introdução

O objetivo de estudo dessa campanha é o entendimento da Função de Coordenação Distribuída(DCF), junto com problema do terminal escondido, o DCF é um método de acesso ao meio implementado pela Media Access Control(MAC), uma subcamada da camada de enlace, sua função por meio do access point (Ap) é coordenar qual estação(sta) deve transmitir, isso é de fudamental importância já que nas telecomunicações gerenciamento e coordenação das transmissões são muito importante.

Podemos dar o exemplo comum na literatura imagine uma sala com muitas pessoas presentes conversando ao mesmo tempo, e você inicia uma conversa com um amigo, mas o ambiente está muito ruidoso não é possivel escutar, com isso você pode tentar gritar ou até desistir da conversa, sendo assim sua comunicação ineficiente, em redes acontece da mesma forma é preciso uma divisão do meio de acesso, para que não ocorra o encotro de pacotes de difrentes transmissões, para que não ocorra colisão e cosequentemente perda de informação, garantindo uma maior eficiência de recursos, com isso é fudamental estudar e solucionar problemas de multiplo acesso.

O Dcf utilizando o carrier sense multiple access collision avoice (cmsa/ca), em que cada termo do seu nome tem um significado:

 * __CS__ : significa carrier sense, refere-se a verificação do dispositivo se o meio está livre antes da transmissão;
  
 * __MA__: multiple access, é referente a divisão do meio onde ocorre o acesso múltiplo;
  
 * __CA__: O collision avoice é pelo fato da sta avisar  sobre as transmissões e seu tempo, de modo a prevenir as colisões. 
 
Funcionando da seguinte forma, se o meio estiver livre após um intervalo de tempo chamdo de DCF Interframe Space(DIFS) uma sta pode transmitir um quadro, entretanto se o meio estiver ocupado, a sta entra em backoff aleatório, em que é um sorteado um tempo aleatório entre zero e janela de contenção(cw), para a próxima uma proxima tentativa de transmissão, depois de esperado DIFS novamente.
O método é ilustrado na figura abaixo :


![fig_figura](FIGS/Figs_tc/dcf.jpeg)

Devido a esse processo, as stas esperam para a transmisão esse metódo de acesso é chamado de acesso por contenção.
Apesar de todo o esforço para evitar colisões tendo a disputa do meio de acesso, elas ainda ocorrem, isso ocasionada por alguns problemas, como o terminal oculto que será explicado abaixo.

### Terminal Oculto 
No terminal oculto uma sta(A) não consegue de detectar outra sta(C) devido ao baixo nível de potência, apesar de ambas estarem ao alcance da sta(B).Com isso podem tentar a transmissão ao mesmo tempo para a sta(B), ocorrendo assim colisões. 

Figura abaixo ilustra o terminal oculto :

![fig_figura](FIGS/Figs_tc/terminal_oculto.png)

## Implementação No ns-3

No ns-3 nosso objetivo foi implementar o terminal oculto, e com isso verificar os efeitos causados por esse problema.
O cenário criado constitui de 2 stas e um ap como mostrado na figura abaixo gerada pelo NetAnim :
![fig_figura](FIGS/Figs_tc/cenario.png)

Para nossa simulação o padrão de camada física escolhido foi 802.11ac. Sendo alguns parâmetros configurados mostrado abaixo:

 * __Largura de banda__ : 160 MHz; 
 
 * __Taxa oferecida Downlink__ :800Mbps ;
 
 * __Taxa oferecida Uplink__ :400Mbps ;
 
 * __Algoritmo de adaptação de taxa__ :MinstrelHtWifiManager ;
 
 * __Tempo de simulação__  : 400 segundos.
 
 ![parametros setados no codigo ](campanha1/FIGS/Figs_tc/parametroscerto.png)
  
O modelo de propagação utilizado é mostrado abaixo, e mais informações podem ser encontradas na [documentação]( https://www.nsnam.org/doxygen/classns3_1_1_log_distance_propagation_loss_model.html.).

![fig_figura](FIGS/Figs_tc/model_propagacao.png)

 
Depois disso usando a classe PositionAlloc o ap foi posicionado no centro, na coordenada (0,0,0) , a sta1 em (-5,0,0) e a sta2 (xd,20,0).
Em que a xd é uma distância variavel na simulação por meio de uma varialvel global, com um valor inicial de 1m, depois o valor 5m, e a partir daí variando em 5 a até chegar ao valor maximo 50 m.

![fig_figura](FIGS/Figs_tc/position_alloc.png)


Com as posições definidas foram configurados os as mobilidade de cada nó da rede, o ap e sta1 permaneceu sem mobilidade tendo sua posição constante utilizando o __ns3::ConstantPositionMobilityModel__, já sta2 além da variação em x na simulação, também realiza um movimento semi-circular se afastando do Ap, sendo realizado por meio da classe __ns3::RandomWalk2MobilityModel__, sendo os valores de seus atributos como direção, velocidade e tempo  mostrado abaixo:  


![fig_figura](FIGS/Figs_tc/mobilidadeap.png)


![fig_figura](FIGS/Figs_tc/mobilidadesta1.png)


![fig_figura](FIGS/Figs_tc/mobilidadesta2.png)
