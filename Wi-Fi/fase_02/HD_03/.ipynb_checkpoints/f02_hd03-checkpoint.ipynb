# Fase II: Campanhas de Investigação com análise de desempenho do Wi-Fi

## Hands-On 03: Campanha 02

### Cenário 01: Posição Constante
* Padrão: 802.11ac;
* Largura de canal: 80MHz;
* Frequência de operação: 5GHz;
* Taxa oferecida: 1.5Mbps (Downlink) e 750Kbps (Uplink);
* Número de STAs: 1, 2, 3, 4, 5, 6, 7, 8, 9 e 10;
* Tempo de simulação: 500 segundos;
* Simulation runs: 1000;
* STAs parados.

Tendo como base o script [rate-adaptation-distance.cc](https://www.nsnam.org/docs/release/3.29/doxygen/rate-adaptation-distance_8cc.html), disponibilizado nos exemplos wireless do ns-3 (*~/ns-3.29/examples/wireless*), e as orientações passadas no [Hands-On 1.5: FlowMonitor, campanhas e plotagem de gráficos](https://nbviewer.jupyter.org/github/vicentesousa/ns3-intermediate-training/blob/master/fase_02/HD_01-5/f02_hd01-5.ipynb) fizemos as alterações necessárias para a construção do primeiro cenário.

A principal alteração no script em relação aos anteriores foi a forma de como a posição das STAs se comportam, no nosso caso elas possuem posições fixas, para isso foi necessário algumas modificações no _PositionAllocator_ das estações. Caso o leitor deseje observar o resultado dessas alterações, o codigo finalizado pode ser encontrado em (*~/ns3-intermediate-training/fase_02/HD_04/CODES*).

### Resultados

A análise de desempenho foi realizada para três tipos de algoritmos de adaptação diferentes, suportados pelo **802.11ac**: 

* IdealWifiManager;
* ConstantRateWifiManager;
* MinstrelHtWifiManager.

São apresentados três plots distintos para Downlink e Uplink: __Throughput vs Nº de STAs__ , **Packet Loss Ratio vs Nº de STAs** e **User Satisfaction vs Nº de STAs**.

### Downlink
#### Throughput
![ThroughputDL](./FIGS/constant/Wifi_nomobility_campaign_TputDL.png)

O throughput oferecido na rede é afetado com o aumento de estações no canal, mas a taxa se comporta de maneiras diferentes de acordo com o algoritmo de adaptação de taxa. O **MinstrelHtWifiManager** possuiu a melhor performance entre os três, obtendo uma mudança brusca na taxa com 9 estações no canal. O **ConstantRateWifiManager** com apenas 3 estações no canal já teve uma redução drástica na taxa ofereciada.
#### Packet Loss Ratio
![PLRDL](./FIGS/constant/Wifi_nomobility_campaign_PLRDL.png)

O gráfico que descreve o comportamento da *Packet Loss Ratio* é complementar ao do throughput, onde a PLR do **MinstrelHtWifiManager** só foi alterada em um canal mais populoso diferentemente do **ConstantRateWifiManager** que com um mínimo aumento de estações já perdia cerca de 40% dos pacotes recebidos.
#### User Satisfaction
![USDL](./FIGS/constant/Wifi_nomobility_campaign_USDL.png)

Para a métrica de User Satisfaction, o critério de taxa utilizado foi 10% da taxa de pico oferecida, nesse caso, 150 Kbps. Validando os resultados obtidos nos outros plots, o algoritmo **MinstrelHtWifiManager** obteve cerca de 100% de satisfação de usuário mesmo com 10 estações no canal, já o **ConstantRateWifiManager** em um canal muito populoso ficou em um patamar de 20% de satisfação.

* Infelizmente a análise do **IdealWifiManager** nos três plots foi afetada devido a algum bug presente no simulador, mas sem uma explicação concreta.

### Uplink

#### Throughput
![ThroughputUL](./FIGS/constant/Wifi_nomobility_campaign_TputUL.png)
Para o Uplink os resultados seguiram a tendência do Downlink, o **MinstrelHtWifiManager** obteve melhores resultados comparados ao **ConstantRateWifiManager**. A diferença entre a análise obtida pelo Downlink foi que os resultados dos dois algoritmos foram semelhantes até 3 estações no canal, onde o trhoughput final do primeiro algoritmo no pior caso (10 estações) terminou em cerca de 700 Kbps e o segundo por volta de 400 Kbps.
#### Packet Loss Ratio
![PLRUL](./FIGS/constant/Wifi_nomobility_campaign_PLRUL.png)
Na análise do PLR para o Uplink, apesar da diferença entre os valores obtidos entre os dois algoritmos foi interessante notar que a PLR geral nesse caso não passou de 20% dos pacotes perdidos no pior caso. O algoritmo continuou com uma melhor performance **MinstrelHtWifiManager**, onde no pior caso de análise chegou a 10% dos pacotes perdidos, enquanto o **ConstantRateWifiManager** ficou em um patamar de 16%.
#### User Satisfaction
![USUL](./FIGS/constant/Wifi_nomobility_campaign_USUL.png)
No quesito de satisfação de usuário, o **MinstrelHtWifiManager** obteve uma curva mais suave que o **ConstantRateWifiManager**, onde o primeiro teve cerca de 50% de usuários satisfeitos no pior caso analisado enquanto o segundo obteve resultados piores, com apenas 20% de usuários satisfeitos no caso mais crítico.

* Infelizmente a análise do **IdealWifiManager** nos três plots foi afetada devido a algum bug presente no simulador, mas sem uma explicação concreta.

### Cenário 02: Posição Aleátoria

* Padrão: 802.11ac;
* Largura de canal: 80MHz;
* Frequência de operação: 5GHz;
* Taxa oferecida: 1.5Mbps (Downlink) e 750Kbps (Uplink);
* Número de STAs: 1, 2, 3, 4, 5, 6, 7, 8, 9 e 10;
* Tempo de simulação: 500 segundos;
* Simulation runs: 1000;
* STAs parados e posicionados aleatóriamente.

Tendo como base o script [rate-adaptation-distance.cc](https://www.nsnam.org/docs/release/3.29/doxygen/rate-adaptation-distance_8cc.html), disponibilizado nos exemplos wireless do ns-3 (*~/ns-3.29/examples/wireless*), e as orientações passadas no [Hands-On 1.5: FlowMonitor, campanhas e plotagem de gráficos](https://nbviewer.jupyter.org/github/vicentesousa/ns3-intermediate-training/blob/master/fase_02/HD_01-5/f02_hd01-5.ipynb) fizemos as alterações necessárias para a construção do segundo cenário.

A principal alteração no script em relação ao anteriore foi a forma de como a posição das STAs se comportam, no segundo caso elas possuem posições fixas que são selecionadas aleatóriamente ao redor do AP, para isso foram necessárias algumas modificações no _PositionAllocator_ das estações. Caso o leitor deseje observar o resultado dessas alterações, o codigo finalizado pode ser encontrado em (*~/ns3-intermediate-training/fase_02/HD_04/CODES*).


## Resultados 
#### Downlink Throughput
Começamos a análise dos resultados com o gráfico do Throughput em função do número de STAs que pode ser visto abaixo:

![fig_figura](./FIGS/Aleatoria/T-put_dl.png)

Pode ser observado que a taxa inicial oferecida está dentro do esperado de 1.5 Mbps, para os algoritmos  MinstrelHtWifiManager e para o ConstanteRateWifiManager, a partir daí devido as característica de se adpatar ao canal do algoritmo MinstrelHtWifiManager manteve sua taxa recebida quase constante com um suave declinio com o aumento do número de estações, já o ConstanteRateWifiManager a partir de duas estações teve um decaimento alto chegando ao valor muito baixo de taxa, o comportamento do ideal seguiu inconclusivo durante a simulação.

#### Uplink Throughput
A análise para o uplink a taxa incial oferecida também está dentro do esperado de 750 kbps,o algortimo MINSTRELHTwifiManeger manteve seu comportamento com uma taxa com pouca variação da inicial, matendo um bom desempenho,o ConstanteRateWifiManager obteve uma melhora em seu desempenho decaindo a partir da 4 estação,chegando um taxa mínima de aproximadamente 350 kbps.

![fig_figura](./FIGS/Aleatoria/T-put_ul.png) 

#### Downlink Packet Loss Ratio

A próxima análise é em relação entre a perda de pacotes, em que o desempenho dos algoritmos foi semelhante as análises anteriores, tendo o MinstrelHtWifiManager com o melhor desempenho devido a baixa perda de de pacotes  independente do numero de STAs, já o ConstanteRateWifiManager apresenta um bom resultado até as duas primeiras STAs a partir daí sofre um aumento considerável, até chegar aos aproximadamente 90% de perda com as 10 STAs.

![fig_figura](./FIGS/Aleatoria/PLR_dl.png)


#### Uplink Packet Loss Ratio

Para uplink a perda de pacotes manteve o mesmo comportamento para o MinstrelHtWifiManager e para o algoritmo ConstanteRateWifiManager apesar de apresentar um baixo desempenho obteve uma melhora considerável tendo uma perda máxima de 16%.

![fig_figura](./FIGS/Aleatoria/PLR_ul.png)

####  Downlink  User Satisfaction

A análise é para User staisfaction com o critério de 10% da taxa observamos novamente o MinstrelHtWifiManager apresenta um bom desempenho, quase 100% de satisfação do usuário para quase todas as STAs, já o ConstanteRateWifiManager como no decorrer de todas as análises apresenta um bom desempenho inicial mais assim que é aumentado o numero de estações começa a decair tendo um nível mínimo de satisfação do usuário de quase 65%.

![fig_figura](./FIGS/Aleatoria/Us_dl.png)

#### Uplink  User Satisfaction

Para uplink a mesma análise do dowlink pode ser feita, o comportamento MinstrelHtWifiManager não muda, e o do ConstanteRateWifiManager sofre uma pequena melhora começando seu decaimento em 4 STAs em vez de 3.

![fig_figura](./FIGS/Aleatoria/Us_ul.png)

### Cenário 03: Posição Aleatória + Mobilidade

* Padrão: 802.11ac;
* Largura de canal: 80MHz;
* Frequência de operação: 5GHz;
* Taxa oferecida: 1.5Mbps (Downlink) e 750Kbps (Uplink);
* Número de STAs: 1, 2, 3, 4, 5, 6, 7, 8, 9 e 10;
* Tempo de simulação: 500 segundos;
* Simulation runs: 1000;
* STAs posicionados aleatóriamente com mobilidade.

Tendo como base o script [rate-adaptation-distance.cc](https://www.nsnam.org/docs/release/3.29/doxygen/rate-adaptation-distance_8cc.html), disponibilizado nos exemplos wireless do ns-3 (*~/ns-3.29/examples/wireless*), e as orientações passadas no [Hands-On 1.5: FlowMonitor, campanhas e plotagem de gráficos](https://nbviewer.jupyter.org/github/vicentesousa/ns3-intermediate-training/blob/master/fase_02/HD_01-5/f02_hd01-5.ipynb) fizemos as alterações necessárias para a construção do terceiro cenário.

A principal alteração no script em relação aos anteriores foi a forma de como a posição das STAs se comportam, no nosso caso elas são posicionadas aleatóriamente e também possuem mobilidade, ou seja, as estações se movimentam no grid. Para satisfazer as necessidades do cenário, foram necessárias algumas modificações no _PositionAllocator_ e _MobilityModel_ das estações. Caso o leitor deseje observar o resultado dessas alterações, o codigo finalizado pode ser encontrado em (*~/ns3-intermediate-training/fase_02/HD_04/CODES*).

### Resultados

A análise de desempenho foi realizada para três tipos de algoritmos de adaptação diferentes, suportados pelo **802.11ac**: 

* IdealWifiManager;
* ConstantRateWifiManager;
* MinstrelHtWifiManager.

São apresentados três plots distintos para Downlink e Uplink: __Throughput vs Nº de STAs__ , **Packet Loss Ratio vs Nº de STAs** e **User Satisfaction vs Nº de STAs**.

### Downlink
#### Throughput
![ThroughputDL](./FIGS/mobility/Wifi_mobility_campaign_TputDL.png)
#### Packet Loss Ratio
![PLRDL](./FIGS/mobility/Wifi_mobility_campaign_PLRDL.png)
#### User Satisfaction
![USDL](./FIGS/mobility/Wifi_mobility_campaign_USDL.png)

Os plots resultantes das simulações mostraram que o **MinstrelHtWifiManager** obteve a melhor performance entre os três algoritmos de adaptação de taxa. Mesmo com a mobilidade no grid seu throughput oferecido só teve uma redução com a presença de estações na mesma rede, e no pior caso (10 STAs) a taxa continuou muito próxima da oferecida com cerca de 20% dos pacotes perdidos e a melhor satisfação. O **ConstantRateWifiManager**, com apenas três STAs presentes na rede já sofre uma perda considerável no throughput, obtendo o pior resultado no cenário mais populoso, chegando a 100% de perda de pacotes e menos de 30% de usuários satisfeitos. Em relação ao **IdealWifiManager** o seu throughput já começa em um patamar menor, comparado aos outros dois algoritmos, mas em um cenário mais populoso ainda obteve melhores resultados que o algoritmo anterior com cerca de 40% de pacotes perdidos e a mesma quantidade de usuários satisfeitos.
### Uplink
#### Throughput
![ThroughputUL](./FIGS/mobility/Wifi_mobility_campaign_TputUL.png)
#### Packet Loss Ratio
![PLRUL](./FIGS/mobility/Wifi_mobility_campaign_PLRUL.png)
#### User Satisfaction
![USDL](./FIGS/mobility/Wifi_mobility_campaign_USUL.png)

Os gráficos gerados para o uplink confirmam a melhor performance do **MinstrelHtWifiManager**. Mesmo com uma taxa ofereceida menor a taxa oferecida para 10 STAs na rede continuou próxima da oferecida, com cerca de 10% de perda de pacotes e 50% de satisfação. Para o **ConstantRateWifiManager** começa com performance similar ao algoritmo anterior, mas cai a partir da presença de 5 STAs, terminando com cerca de 20% de perda de pacotes e uma péssima satisfação de usuários. Ao utilizar o **IdealWifiManager** seu throughput foi o menor entre os três, com uma alta perda de pacotes para poucas estações, e, mesmo com uma vazão útil menor que a do **ConstantRateWifiManager** obteve resultados similares com uma grande quantidade de estações na rede, cerca de 20% de perda de PLR e menos de 10% de satisfação.
