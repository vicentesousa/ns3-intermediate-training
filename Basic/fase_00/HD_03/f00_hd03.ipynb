# Hands-on 3: Instalando o ns-3 e integrando com o Eclipse

# Objetivos
As metas desse tutorial são ajudar o usuário a:
- Instalar o ns3 e dependências;
- Integrar o ns3 com o eclipse.

# Instalar o ns-3


Foi criado um script que chama todas as dependências em sequência, incluindo a instalação do Eclipse. Para executá-lo siga uma dos seguintes maneiras:

E script está disponível [aqui](./CODE/ns3_full_installation.sh).


**Maneira 1**: 

1) Mudá-lo para executável: sudo chmod +x  ns3_full_installation.sh

2) Executá-lo: ./ns3_full_installation.sh

**Maneira 2**: 

1) bash ./ns3_full_installation.sh

---------------

O scritp instalará o a versão ns-3.29 na seguinte pasta: ~/ns-3/ns-allinone-3.29/ns3-3.29

Para compilar o ns-3 posteriormente (caso mude o código), use:

Para debug: CXXFLAGS="-Wall -g -O0“ ./waf --build-profile=debug --enable-examples --enable-tests configure

Para código otimizado: CXXFLAGS="-Wall -g -O0“ ./waf --build-profile=optimized --enable-examples --enable-tests configure


# **Eclipse:**

Caso queira utilizar uma versão diferente do eclipse, no site (https://www.eclipse.org/downloads/packages/), há diferentes versões da IDE. Fique atento, pois algumas configurações pontuais não listadas aqui podem ser necessárias para integração com o ns3.

# Integração com o Eclipse


Os passos para integração com o eclipse só podem ser feitos se a instalação ocorreu sem maiores problemas. O Eclipse é uma ferramente de desenvolvimento que tem suporte a diversas liguagens, dentre elas c++, a linguagem de programação utilizada pelo ns-3. Segue os passos para integração do ns-3 com o Eclipse:


## Passo 1: Crie um novo projeto c/c++.

![eclipse1](./FIGS/eclipse1.png)

## Passo 2: Crie um projeto novo no eclipse, coloque o folder do ns3 como principal.

![eclipse2](./FIGS/eclipse2.png)

### Passo 2.1: Depois de criado, o projeto do ns3 deve aparecer na lista dos projetos.

![eclipse4](./FIGS/eclipse4.png)

## Passo 3: Com o botão direito, vá em "properties". 

![eclipse3](./FIGS/eclipse3.png)

## Passo 4: Nas propriedades do projeto, em C/C++ build adicionar:

Build command: path_ns3/waf

Build director: path_ns3

![eclipse5](./FIGS/eclipse5-1.png)

### Passo 4.1 Na aba "Behavior", mude o campo "Build (incremental build)" para build.

![eclipse6](./FIGS/eclipse6.png)

## Passo 5: Utilize o comando "Build Project".

![eclipse7](./FIGS/eclipse7.png)

## Passo 6: Crie um perfil de simulação.


### Passo 6.1: Crie o perfil de simulação.

1) Em "Run Configurations" ou "Debug Configurations".

![eclipse8](./FIGS/eclipse8.png)

### Passo 6.2: Crie o perfil.

![eclipse9](./FIGS/eclipse9.png)

### Passo 6.3: Necessário definir: 

1) o nome do projeto referente ao ns3; 

2) o script que será usado na simulação (clicar em Search Project...);

![eclipse10](./FIGS/eclipse10.png)


### Passo 6.4: Na aba "Environment":

1) Adicionar a variável: LD_LIBRARY_PATH, cujo valor é: 

path_ns3/build/ 

ou 

path_ns3/build/lib (vai depender da versão do ns3);


![eclipse12](./FIGS/eclipse11-1.png)

--------------

Após a realização de todos esses passos será possível utilizar o eclipse como plataforma de programação do ns-3.

## Oberservações:

Caso faça novas instalações do ns-3 e queira habilitar o debug no Eclipse, utilize a linha no terminal para cada nova instalação do ns3:

1) sudo CXXFLAGS="-Wall -g -O0" ./waf --build-profile=debug --enable-examples --enable-tests configure
